```python
#import the libraries
import numpy as np
import matplotlib.pyplot as plt

#library that contains the function k-means
from sklearn.cluster import KMeans

#library that contains our dataset
from sklearn.datasets import load_digits
```

```python
#load the digit images

digits = load_digits()
data = digits.data
print(data.shape)

#To improve the visualization, invert the colors
data = 255-data
```

```python
#finx the seed to obtain the initial centroids so that the results are repeatable.
np.random.seed(1)

#we have 10 diff digits(0-9), so the images can be grouped into 10 clusters
n=10

#classify the data with k-means
kmeans = KMeans(n_clusters=n,init='random')
kmeans.fit(data)
Z = kmeans.predict(data)
```

```python
#plot the resulting clusters
for i in range(0,n):

    row = np.where(Z==i)[0]       # row in Z for elements of cluster i
    num = row.shape[0]            #  number of elements for each cluster
    r = int(np.floor(num/10.))    # number of rows in the figure of the cluster

    print("cluster "+str(i))
    print(str(num)+" elements")

    plt.figure(figsize=(10,10))
    for k in range(0, num):
        plt.subplot(r+1, 10, k+1)
        image = data[row[k], ]
        image = image.reshape(8, 8)
        plt.imshow(image, cmap='gray')
        plt.axis('off')
    plt.show()
```
